```python
# Create a new dataframe with the keywords and the announces
keywords_df = pd.DataFrame(scores, columns=words)
keywords_df['URL'] = df['URL']  # The url is the key to map each announce

print(keywords_df)
```

```python
# Function to extract the most important keywords from each announce
def get_top_keywords(row):
    top_indices = row.nlargest(3).index  # Obtain the 3 parole più significative
    return ', '.join(top_indices)
```

```python
# Apply the function to each announce
df['Top Keywords'] = keywords_df.drop('URL', axis=1).apply(get_top_keywords, axis=1)
```
